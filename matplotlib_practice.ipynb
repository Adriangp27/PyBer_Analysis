# Creating Scatter Plot using MATLAB Method

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(x_axis, y_axis)
```

```python
# Another alternative method for scatter plot
plt.plot(y_axis, x_axis, 'o', color = 'red', label = 'Chicago')

# Annotates the scatter plot
plt.xlabel('Fare($)')
plt.ylabel('Date')
plt.legend()

# Inverts the y-axis so january is on top
plt.gca().invert_yaxis()
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Modifying bubble chart by using existing scatter plot code.
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# New list comprehension technique to multipyjeach data point in the y-axis.
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
#create the bubble chart
fig, ax = plt.subplots()

ax.scatter(y_axis, x_axis, s = [i * 5 for i in y_axis] , c = 'skyblue', alpha = 0.8, edgecolor = 'black', linewidth = 2, label = 'Boston')

#anontate the chart
ax.set_xlabel('Fare ($)')
ax.set_ylabel('Date')
ax.set_title('PyBer Fare by Month')
ax.legend()

#inverts y axis
ax.invert_yaxis()

#limits x axis data
ax.set_xlim(0,50)
```

# Adding Error bars and Changing Ticks
Error Bars to a line chart using Matlab interface

```python
%matplotlib inline
```

```python
# Import dependencies.
import matplotlib.pyplot as plt
import statistics
```

```python
# Setting the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Setting the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Getting the standard deviation of the values in the y-axis.
stdev = statistics.stdev(y_axis)
stdev
```

```python
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# Using Object-Oriented Approach.
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
# Adding Error Bars to Bar Chart
plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# Creating same graph using Object-Oriented Approach
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
# Changing Major Ticks
import numpy as np
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()
```

```python
# Same graph but using object-oriented approach
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))
plt.show()
```

```python
from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
